```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x,y = sp.symbols('x y')

f = sp.exp(x) * sp.sin(y) + y ** 3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("df/dx is "+str(df_dx)+"\ndf/dy is "+str(df_dy))

g = sp.log(x ** 2 + y ** 2)
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

print("dg/dx is "+str(dg_dx)+"\ndg/dy is "+str(dg_dy))
print("At the point (1,-1), the gradient vector is (1,-1). The magnitude is sqrt(2).")
h = sp.log(x**2+y**2)
dh_dx = sp.diff(h,x)
dh_dy = sp.diff(h,y)
d2h_dx2 = sp.diff(dh_dx,x)
d2h_dy2 = sp.diff(dh_dy,y)
d2h_dydx = sp.diff(dh_dx,y)
d2h_dxdy = sp.diff(dh_dy,x)

print("dh/dx is "+str(dh_dx)+"\ndh/dy is "+str(dh_dy)+"\nd2h/dx2 is "+str(d2h_dx2)+"\nd2h/dy2 is "+str(d2h_dy2)+"\nd2h/dydx is "+str(d2h_dydx)+"\nd2h/dxdy is "+str(d2h_dxdy))
print("The mixed second order partial derivatives are the same.")
# def f_grad(x, y):
#     dfdx = np.exp(x) * np.sin(y)
#     dfdy = np.exp(x) * np.cos(y) + 3 * y ** 2
#     return dfdx, dfdy
# 
# 
# def g(x, y):
#     return np.log(x ** 2 + y ** 2)
# 
# 
# def g_grad(x, y):
#     dgdx = 2 * x * y + y ** 2
#     dgdy = x ** 2 + 2 * x * y
#     return dgdx, dgdy  
# 

# gradient = g_grad(1, -1)
# print("The magnitude of the gradient of g(x,y) at (1,-1) is " + str(np.sqrt(gradient[0] ** 2 + gradient[1] ** 2)))
# 
# 
# def h(x, y):
#     return np.log(x ** 2 + y ** 2)
# 
# 
# def h_second_pd(x, y):
#     # d2hdx2 = (2 * (x ** 2 + y ** 2) - 4 * x ** 2) / (x ** 2 + y ** 2) ** 2
#     # d2hdy2 = (2 * (x ** 2 + y ** 2) - 4 * y ** 2) / (x ** 2 + y ** 2) ** 2
#     # d2hdydx = -4 * x * y / ((x ** 2 + y ** 2) ** 2)
#     
#     
#     return d2hdx2, d2hdy2, d2hdydx
# 
# 
# # The mixed second order partial derivatives are the same
# 
# def j(x, y):
#     return x ** 3 - 3 * x * y + y ** 3


# first function j

x, y = sp.symbols('x y')
j = x ** 3 - 3 * x * y + y ** 3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x,y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# second function k

k = x**2+y**3
k_func = sp.lambdify((x, y), k, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contour(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x,y) = x^2 + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
